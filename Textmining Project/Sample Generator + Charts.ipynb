```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### **GENERATOR PRÃ“BY**

```python
df = pd.read_csv("Comments.csv", sep=';')
```

```python
date = "2020-12-10"
before = df['comment_date'] < date
after = df['comment_date'] >= date
```

```python
df_before = df.loc[before]
df_before = df_before["comment_body"].sample(5000)
df_before.to_csv("comment_body_before.csv", index = False)
```

```python
df_after = df.loc[after]
df_after = df_after["comment_body"].sample(5000)
df_after.to_csv("comment_body_after.csv", index = False)
```
